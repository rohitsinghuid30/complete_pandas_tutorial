# String Filters

# 1. Find athelete born in cities that starts with vowel

```python
vowel_cities=bios[bios["born_city"].str.contains(r'^[AEIOUaeiou]', na=False)]
```

```python
vowel_cities
```

# 2. Find athelete with names that contains exactly two vowel

```python
two_vowel=bios[bios['name'].str.contains(r'^[^AEIOUaeiou]*[AEIOUaeiou][^AEIOUaeiou]*[AEIOUaeiou][^AEIOUaeiou]*$',na=False)]
```

```python
two_vowel
```

# 3. Find atheletes with names that have repeted consecutive letters (e.g "Arron", "Emmmett")

```python
repeated_letters = bios[bios['name'].str.contains(r'(.)\l', na=False)]
```
